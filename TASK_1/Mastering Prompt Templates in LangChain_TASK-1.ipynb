```python
# Import from langchain_core to ensure compatibility with latest versions
from langchain_core.prompts import (
    PromptTemplate, 
    ChatPromptTemplate, 
    SystemMessagePromptTemplate, 
    HumanMessagePromptTemplate
)

# ==========================================
# Task 1: Replace Hardcoded Prompts (10%)
# ==========================================
template_v1 = "Explain {topic} in simple terms for beginners"
prompt_template = PromptTemplate(input_variables=["topic"], template=template_v1)

print("--- Task 1: PromptTemplate Output ---")
print(prompt_template.format(topic="Quantum Physics"))
print("\n")


# ==========================================
# Task 2: Multi-Input Prompt System (15%)
# ==========================================
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print("--- Task 2: Multi-Input Test Cases ---")
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for case in test_cases:
    print(multi_input_template.format(**case))
print("\n")


# ==========================================
# Task 3: Prompt Variations Engine (15%)
# ==========================================
variations = {
    "teaching": PromptTemplate.from_template("Explain {topic} clearly step by step"),
    "interview": PromptTemplate.from_template("Ask 3 questions about {topic}"),
    "storytelling": PromptTemplate.from_template("Explain {topic} as a story")
}

print("--- Task 3: Variations for Machine Learning ---")
topic_ml = "Machine Learning"
for key, tmpl in variations.items():
    print(f"{key.capitalize()}: {tmpl.format(topic=topic_ml)}")
print("\n")


# ==========================================
# Task 4: ChatPromptTemplate System (15%)
# ==========================================
def get_chat_prompt(role, topic):
    roles = {
        "teacher": "You are a patient teacher who simplifies complex concepts.",
        "interviewer": "You are a rigorous tech interviewer looking for deep technical knowledge.",
        "motivator": "You are an inspiring coach who encourages growth and persistence."
    }
    
    system_msg = SystemMessagePromptTemplate.from_template(roles.get(role, "You are a helpful assistant."))
    human_msg = HumanMessagePromptTemplate.from_template("Topic: {topic}")
    
    chat_prompt = ChatPromptTemplate.from_messages([system_msg, human_msg])
    # Returns formatted messages
    return chat_prompt.format_prompt(topic=topic).to_messages()

print("--- Task 4: ChatPromptTemplate (Teacher Role) ---")
print(get_chat_prompt("teacher", "Neural Networks"))
print("\n")


# ==========================================
# Task 5: Input Validation Layer (10%)
# ==========================================
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]
    
    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "formal"
    
    return final_audience, final_tone


# ==========================================
# Task 6: Prompt Generator App (15%)
# ==========================================
def generate_prompt(topic, audience, tone, style):
    v_audience, v_tone = validate_inputs(audience, tone)
    
    style_map = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style.",
        "interview": "Generate questions about {topic} for {audience} in a {tone} interview style.",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style."
    }
    
    selected_tmpl = style_map.get(style, style_map["teaching"])
    final_template = PromptTemplate.from_template(selected_tmpl)
    
    return final_template.format(topic=topic, audience=v_audience, tone=v_tone)

print("--- Task 6: Prompt Generator App Output ---")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print("\n")


# ==========================================
# Task 7: Template Reusability Test (10%)
# ==========================================
reusable_template = PromptTemplate.from_template("Guide to {topic} for {audience} ({tone} version)")

print("--- Task 7: Reusability Test (1 Template, 5 Inputs) ---")
inputs = [
    {"topic": "Cooking", "audience": "amateurs", "tone": "casual"},
    {"topic": "SEO", "audience": "marketers", "tone": "formal"},
    {"topic": "Yoga", "audience": "seniors", "tone": "gentle"},
    {"topic": "Cybersecurity", "audience": "employees", "tone": "urgent"},
    {"topic": "Space Travel", "audience": "tourists", "tone": "fun"}
]

for item in inputs:
    print(reusable_template.format(**item))
```
